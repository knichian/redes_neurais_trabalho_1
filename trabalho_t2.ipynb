```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score, f1_score, recall_score 
from sklearn.preprocessing import LabelEncoder

from itertools import product
```

```python
# separar features de targers
X = df_filtered.drop(columns='diagnosis')
y = LabelEncoder().fit_transform(df_filtered['diagnosis'].astype(str))
# y = df_filtered['diagnosis']
```

```python
# divisÃ£o do dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
def build_model( hidden1, hidden2, activation_mid, activation_end, learning_rate ):
  attr_amount = 21
  
  model = Sequential([
    Dense(hidden1, activation=activation_mid, input_shape=(attr_amount,)),
    Dense(hidden2, activation=activation_mid),
    Dense(2, activation=activation_end)
  ])
  
  model.compile(optimizer=Adam(learning_rate=learning_rate),
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
  
  return model
```

```python
# create grid search
# make the grid search decide "classifier" parameters with it's own arguments
# make it save the model
# make it save the result of all metrics he requested

# epochs = 200
# batch_size = 32

# from ast import arg


def grid_search( hidden1_vals, hidden2_vals, activation_mid_vals, activation_end_vals, lr_vals, epochs_vals, batch_size_vals ):
  
  best_result = {
      'params': 0,
      'score_accuracy': 0,
      'score_precision': 0,
      'score_f1': 0,
      'score_recall': 0,
    }
  
  for hidden1, hidden2, activation_mid, activation_end, learning_rate, epochs, batch_size in product( hidden1_vals, hidden2_vals, activation_mid_vals, activation_end_vals, lr_vals, epochs_vals, batch_size_vals ):  
    model = build_model( hidden1, hidden2, activation_mid, activation_end, learning_rate )
    model_history = model.fit( X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=True )
    y_pred = np.argmax(model.predict(X_test), axis=1)
    
    params = (hidden1, hidden2, activation_mid, activation_end, learning_rate, epochs, batch_size)
    
    score_accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
    score_precision = precision_score(y_true=y_test, y_pred=y_pred)
    score_f1 = f1_score(y_true=y_test, y_pred=y_pred)
    score_recall = recall_score(y_true=y_test, y_pred=y_pred)
    
    model_result = {
      'params': params,
      'score_accuracy': score_accuracy,
      'score_precision': score_precision,
      'score_f1': score_f1,
      'score_recall': score_recall,
    }
    
    display(model_result)
    
    if ( model_result['score_accuracy'] > best_result['score_accuracy'] ):
      print('*** new_best ***')
      display(model_result)
      best_result = model_result
    
  
  print('*** final_best ***')
  display(best_result)
  
  return best_result
```

```python
# testando funÃ§Ã£o
hidden1_vals = [ 7 ]
hidden2_vals = [ 6 ]
activation_mid_vals = [ 'relu' ]
activation_end_vals = [ 'sigmoid' ]
lr_vals = [ 0.01 ]
epochs_vals = [ 10 ] 
batch_size_vals = [ 32 ]

best_result = grid_search( hidden1_vals, hidden2_vals, activation_mid_vals, activation_end_vals, lr_vals, epochs_vals, batch_size_vals )
```

```python
# Definindo os hiperparÃ¢metros a testar

hidden1_vals = [ 21, 7 ]
hidden2_vals = [ 6, 7 ]
activation_mid_vals = [ 'relu' ]
activation_end_vals = [ 'sigmoid', 'tanh']
lr_vals = [ 0.001, 0.01 ]
epochs_vals = [ 200 ] 
batch_size_vals = [ 32 ]

best_result = grid_search( hidden1_vals, hidden2_vals, activation_mid_vals, activation_end_vals, lr_vals, epochs_vals, batch_size_vals )
```
